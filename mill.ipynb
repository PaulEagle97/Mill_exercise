```python
# Set up environment
import pandas as pd
import numpy as np
import seaborn as sns

from scipy.signal import find_peaks
from matplotlib import pyplot as plt
```

## Data Retrieval

```python
# Read the data
FNAME =  './mill.csv'
# Convert to DataFrame format
df = pd.read_csv(FNAME, sep = ';', dtype={'Sorte':object}, parse_dates=['Datum'])
df.columns = ['Grade', 'ZM', 'Date', 'Uptime', 'Total Energy', 'Output', 'kWh/t', 't/h']
```

## Data Inspection

```python
df.info()
```

And there appear to be invalid values in the table:

```python
df.describe()
```

## Data Cleaning

### Dealing with Uptime

There are **"Uptime"** entries which appear to be approaching zero:

```python
df.Uptime.plot(kind='hist', bins=40)
```

```python
temp = df.loc[df['Uptime'] <= 0.2]
```

```python
temp.Uptime.plot(kind='hist', bins=30)
```

We will discard these entries as invalid:

```python
df = df[df['Uptime'] > 0.2]
```

### Dealing with Output

Some entries of **Output** appear to be negative or approaching zero:

```python
df.Output.plot(kind='hist', bins = 100)
```

```python
df.loc[df['Output'] <= 0]
```

```python
df.Output.plot(kind='hist', bins = 100).set_xlim([df['Output'].min(), 0])
```

We will discard these entries as invalid:

```python
df = df[df['Output'] > 0]
```

### Dealing with Duplicates

```python
# Check for duplicates
duplicate_rows = df[df.duplicated(subset=['Date', 'Grade', 'ZM'], keep=False)].sort_values(['Date'])
duplicate_rows
```

Apparently, there are no duplicates left

#### Plotting **kWh/t**:

```python
sorted_kwht = df.sort_values(by='kWh/t')['kWh/t']
filtered_kwht = sorted_kwht.head(len(sorted_kwht)-9)

# Get histogram values and bin edges
hist_values, bin_edges, _ = plt.hist(filtered_kwht, bins=100)

# Find peaks in the histogram
peaks, _ = find_peaks(hist_values, prominence=40, width=1.5)

# Get x-coordinates of the peaks
peak_x_coordinates = bin_edges[peaks]

plt.plot(hist_values)
plt.plot(peaks, hist_values[peaks], "x")
for x, y in zip(peaks, hist_values[peaks]):
    plt.text(x, y, f'({x}, {y:.0f})', ha='right', va='bottom')

plt.show()
```

```python
filtered_kwht.plot(kind='kde')
plt.show()
```

#### Plotting **t/h**:

```python
sorted_th = df.sort_values(by='t/h')['t/h']
filtered_th = sorted_th.head(len(sorted_th)-5)
# Get histogram values and bin edges
hist_values, bin_edges, _ = plt.hist(filtered_th, bins=100)

# Find peaks in the histogram
peaks, _ = find_peaks(hist_values, prominence=40, width=1.5)

# Get x-coordinates of the peaks
peak_x_coordinates = bin_edges[peaks]

plt.plot(hist_values)
plt.plot(peaks, hist_values[peaks], "x")
for x, y in zip(peaks, hist_values[peaks]):
    plt.text(x, y, f'({x}, {y:.0f})', ha='right', va='bottom')

plt.show()
```

```python
filtered_th.plot(kind='kde')
plt.show()
```

#### Conclusion

As per illustrated by the graphs, we can conclude that the **power consumption efficiency** of the mill is well defined by two modes: **27KWh/t** and **32 KWh/t**.
When it comes to its **productivity**, 3 main modes of operation emerge: **17 t/h**, **42 t/h** and **52 t/h**.

### Relationship Analysis

#### General Overview

```python
sns.pairplot(df, vars=['Date', 'Uptime', 'Total Energy', 'Output'], hue='ZM')
plt.show()
```

As we can see from the heatmap below, all 3 primary indicators are strongly and positively correlated with each other:

```python
df_corr = df[['Uptime', 'Total Energy', 'Output']].corr()
sns.heatmap(df_corr, annot=True, vmin=0, vmax=1)
```

#### **Grade** vs **ZM**

```python
df.groupby('Grade')['ZM'].value_counts().unstack().plot(kind='bar')
plt.ylabel('Occurrences')
plt.show()
```

This graph shows how **Grade** relates to **ZM**. For instance, we can see that **Grade 02** was never produced with **ZMA**.

#### **ZM** vs **t/h**

```python
grouped = df.groupby('ZM')['t/h'].mean()
ax = grouped.plot(kind='bar')
ax.bar_label(ax.containers[0], fmt='{:.1f}')
plt.xlabel('')
plt.ylabel('Mean t/h')
plt.show()
```

The graph above demostrates that **ZMC** is associated with a low productivity of the mill.

#### **kWh/t** vs **t/h**

```python
sns.scatterplot(x='t/h', y='kWh/t', data=df, hue='ZM')
# Limit the x-axis and y-axis range
plt.xlim(0, 100)  # Set x-axis limits
plt.ylim(0, 150)  # Set y-axis limits

# Show the plot
plt.savefig('./scatter_plot_kWht_vs_th.png')
plt.show()
```

As per shown in the graph, we can see that there is correlation between **power efficiency** and **productivity** of the mill. The graph can be approximated by a composite function, consisting of 2 smooth, concave up and decreasing curves with a jump discontinuity at approx. **30 t/h**. The first part of the curve is represented mainly by **ZMC** and then it transitions into **ZMA/ZMB**.
This data alligns well with the principles of economy of scale, which states that the bigger the production, the less resources are spent per unit of production. The jump at the mark **30 t/h** can be explained such that, as the mill increases its output, at some point it requires to switch from the more economic **ZMC** to **ZMA/ZMB** in order to support an increase in production.

#### **Total Energy** vs **Uptime**

```python
sns.scatterplot(x='Uptime', y='Total Energy', data=df, hue='ZM')

# Show the plot
plt.savefig('./scatter_plot_Uptime_vs_TotalEnergy.png')
plt.show()
```

As we can see from the graph, **ZM** directly influences the energy consumption, which in all 3 cases is clearly linear to the time that the mill has spent working. We can compute an overall power of the mill depending on the **ZM**:

```python
df['Power'] = df['Total Energy'] / df['Uptime']
```

```python
filtered_df = df[df['ZM'] == 'ZMC']
zmc_power = filtered_df['Power'].mean()
filtered_df = df[df['ZM'] != 'ZMC']
zma_zmb_power = filtered_df['Power'].mean()
```
